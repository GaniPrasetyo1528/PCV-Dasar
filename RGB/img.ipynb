```python
# menampilkan gambar sederhana menggunakan cv2 + matplotlib
import cv2
import matplotlib.pyplot as plt

# Membuat variabel img untuk menyimpan gambar
# cv2.imread() digunakan untuk membaca gambar
# dan menyimpannya ke dalam variabel img
img = cv2.imread('../IMG/musclecars.jpg')

# Jika menggunakan cv2 untuk menampilkan gambar, maka warnanya akan menjadi BGR
# plt.imshow() akan menampilkan gambar dalam format RGB
plt.imshow(img)
```
