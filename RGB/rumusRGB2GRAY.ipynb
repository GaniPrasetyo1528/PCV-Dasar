```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../IMG/musclecars.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

tinggi, lebar, _ = img_rgb.shape

# gray merupakan array kosong untuk menampung hasil grayscale
# dtype=np.uint8 digunakan agar array gray bertipe unsigned 8-bit integer (0-255), sesuai rentang nilai pixel gambar grayscale
gray = np.zeros((tinggi, lebar), dtype=np.uint8)

for y in range(tinggi):
    for x in range(lebar):
        r, g, b = img_rgb[y, x]   # ambil nilai R, G, B
        # rumus grayscale
        gray[y, x] = int(0.299*r + 0.587*g + 0.114*b)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Asli (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale (Rumus)")
plt.axis("off")

plt.show()
```
