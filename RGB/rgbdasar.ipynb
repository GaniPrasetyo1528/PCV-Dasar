```python
import cv2
import matplotlib.pyplot as plt

# Jika hanya menggunakan cv2.imread, warna defaultnya adalah BGR
img = cv2.imread('../IMG/musclecars.jpg')
# Disini saya menambahkan konversi BGR ke RGB menggunakan cv2.cvtColor (Convert Color) dan menyimpannya ke variabel baru(img_rgb)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan gambar agar terlihat dengan baik menggunakan matplotlib
# plt.imshow() untuk menampilkan gambar beserta arraynya
plt.imshow(img_rgb)
# Title untuk gambar
plt.title('Muscle Cars')
# axis('off') menghilangkan axis pada gambar/meghapus garis kotak pada gambar
plt.axis('off')
# plt.show() menampilkan keseluruhan gambar
plt.show()
```
