```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../IMG/musclecars.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan 2 gambar dalam 1 kanvas
# subplot (baris, kolom, index) mengatur posisi gambar
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Format BGR')
plt.axis('off')
# Note: Jangan menaruh plt.show() setelah code gambar pertama. jika ditaruh padasetiap akhir code gambar, maka kanvas akan terpisah.

plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.title('Format RGB')
plt.axis('off')
plt.show()
```
