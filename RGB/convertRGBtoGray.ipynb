```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../IMG/musclecars.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Disini mengubah rgb ke gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,3,1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_rgb)
plt.title('RGB Image')
plt.axis('off')

plt.subplot(1,3,3)
# Gunakan cmap untuk menampilkan gray karena gray hanya 1 channel sedangkan plt.imshow defaultnya 3 channel
plt.imshow(img_gray, cmap='gray')
plt.title('Gray Image')
plt.axis('off')
plt.show()
```
