```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread("../IMG/musclecars.jpg", cv2.IMREAD_GRAYSCALE)

height, width = img_gray.shape

negative_loop = np.zeros_like(img_gray)

for y in range(height):
    for x in range(width):
        negative_loop[y, x] = 255 - img_gray[y, x]

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap="gray")
plt.title("Asli")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(negative_loop, cmap="gray")
plt.title("Negatif (loop)")
plt.axis("off")

plt.show()
```
