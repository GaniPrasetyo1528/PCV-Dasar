```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../IMG/musclecars.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Citra negatif (langsung kurangi dari 255)
negative_rgb = 255 - img_rgb

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Asli (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(negative_rgb)
plt.title("Negatif (RGB)")
plt.axis("off")

plt.show()
```
