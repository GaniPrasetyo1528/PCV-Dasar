```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_gray = cv2.imread('../IMG/musclecars.jpg', cv2.IMREAD_GRAYSCALE)

Imin, Imax = 100, 200
# tanpa loop
img_clipped = np.clip(img_gray, Imin, Imax)

height, width = img_gray.shape
# np.zeros_like untuk membuat array baru dengan ukuran yang sama
img_clipped_loop = np.zeros_like(img_gray)

# Loop setiap piksel
for y in range(height):
    for x in range(width):
        f = img_gray[y, x]  # intensitas asli
        if f < Imin:
            img_clipped_loop[y, x] = Imin
        elif f > Imax:
            img_clipped_loop[y, x] = Imax
        else:
            img_clipped_loop[y, x] = f

# Tampilkan hasil
plt.subplot(2, 2, 1)
plt.title('Gambar Asli')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Gambar Setelah Clipping (Loop)')
plt.imshow(img_clipped_loop, cmap='gray')
plt.axis('off')

plt.subplot(2, 1, 2)
plt.imshow(img_clipped, cmap='gray')
plt.title('Gambar Setelah Clipping (Tanpa Loop)')
plt.axis('off')
plt.show()
```
