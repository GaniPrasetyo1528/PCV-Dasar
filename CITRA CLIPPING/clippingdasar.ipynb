```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_gray = cv2.imread('../IMG/musclecars.jpg', cv2.IMREAD_GRAYSCALE)

# Tentukan nilai intensitas minimum dan maksimum
Imin, Imax = 100, 200

# Lakukan clipping
img_clipped = np.clip(img_gray, Imin, Imax)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Gambar Asli')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gambar Setelah Clipping')
plt.imshow(img_clipped, cmap='gray')
plt.axis('off')
plt.show()
```
